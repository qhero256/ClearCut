```python
import subprocess
import os
import time
from datetime import datetime

def procesar_un_video(archivo_entrada, margen="0.2s"):
    if not os.path.exists(archivo_entrada):
        print(f"‚ùå No encuentro '{archivo_entrada}'")
        return
    nombre, ext = os.path.splitext(archivo_entrada)
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    archivo_salida = f"{nombre}_EDITADO_{ts}{ext}"
    comando = [
        "auto-editor",
        archivo_entrada,
        "--silent-speed", "0",
        "--margin", margen,
        "-o", archivo_salida
    ]
    print("‚è≥ Eliminando silencios‚Ä¶")
    print(f"üìÇ Entrada: {archivo_entrada}")
    print(f"üíæ Salida:  {archivo_salida}")
    inicio = time.time()
    try:
        subprocess.run(comando, check=True, shell=True)
        fin = time.time()
        print("‚úÖ Proceso completado")
        print(f"‚è±Ô∏è Duraci√≥n: {fin-inicio:.1f}s")
    except subprocess.CalledProcessError as e:
        print("‚ùå Auto-Editor fall√≥:", e)

# Ejec√∫talo as√≠:
procesar_un_video(r"C:\Users\Nombre\Documents\27122025_2016.mp4", margen="0.2s")
```
